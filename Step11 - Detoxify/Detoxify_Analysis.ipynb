```python
import pandas as pd
from detoxify import Detoxify
import numpy as np
```

```python
tsv = pd.read_csv('final.tsv', sep='\t')
tsv
```

```python
tsv.columns
```

```python
tsv.Toxicity = tsv.Toxicity.astype(float).fillna(0.0)
tsv.Severe_Toxicity = tsv.Severe_Toxicity.astype(float).fillna(0.0)
tsv.Obscenity = tsv.Obscenity.astype(float).fillna(0.0)
tsv.Identity_Attack = tsv.Identity_Attack.astype(float).fillna(0.0)
tsv.Insult = tsv.Insult.astype(float).fillna(0.0)
tsv.Threat = tsv.Threat.astype(float).fillna(0.0)
tsv.Sexual_Explicit = tsv.Sexual_Explicit.astype(float).fillna(0.0)
```

```python
# extract valuable columns
df = tsv[['ifHate', 'ifSarcasm', 'rtg_translate', 'Toxicity', 'Severe_Toxicity', 'Obscenity', 'Identity_Attack',
       'Insult', 'Threat', 'Sexual_Explicit']]
```

```python
# create 4 sub-dataframes according to the ifHate and ifSarcasm lable
hate = df[df['ifHate'] == True]
sarcasm = df[df['ifSarcasm'] == True]
nothate = df[df['ifHate'] == False]
notsarcasm = df[df['ifSarcasm'] == False]
```

```python
# stats of rows being labled as hate speech
hate.describe()
```

```python
# setting the threshold to 0.5, any entry with a value >= 0.5 is considered toxic/obscene/insulting/...
# count how many rows are considered problematic by both methods
hate[(hate['Toxicity']>=0.5) | (hate['Severe_Toxicity']>=0.5) | (hate['Obscenity']>=0.5) 
     | (hate['Identity_Attack']>=0.5) | (hate['Insult']>=0.5) | (hate['Threat']>=0.5) 
     | (hate['Sexual_Explicit']>=0.5)].describe()
```

```python
# get number of problematic posts of each category
print('Number of toxic post: '+str(hate[(hate['Toxicity']>=0.5)].shape[0]))
print('Number of severely toxic post: '+str(hate[(hate['Severe_Toxicity']>=0.5)].shape[0]))
print('Number of obscene post: '+str(hate[(hate['Obscenity']>=0.5)].shape[0]))
print('Number of indentity attacking post: '+str(hate[hate['Identity_Attack']>=0.5].shape[0]))
print('Number of insulting post: '+str(hate[(hate['Insult']>=0.5)].shape[0]))
print('Number of threatening post: '+str(hate[(hate['Threat']>=0.5)].shape[0]))
print('Number of sexual explicit post: '+str(hate[(hate['Sexual_Explicit']>=0.5)].shape[0]))
```
