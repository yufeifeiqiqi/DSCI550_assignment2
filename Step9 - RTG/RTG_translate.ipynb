```python
import numpy as np
import pandas as pd
import re
from tika import translate
from tika import language
```

```python
combine = pd.read_csv('combine_lan_detect.csv')
```

```python
def remove_tag(in_str): #remove tag
    left = 0
    new_str = ""
    
    for i in in_str:
        if i == '<':
            left += 1
        elif left == 0:
            new_str += i
        elif i == '>':
            left -= 1
    out_str = ""        
    left2 = 0
    for i in new_str:
        if i == '&':
            left2 += 1
        elif left2 == 0:
            out_str += i
        elif i == ';':
            left2 -= 1
            out_str += ' '
    return(out_str)
```

```python
def find_regular(content_raw): # remove url
    content_raw1 = content_raw.replace('#', ' ').replace('【', '').replace('】', '').replace('<p>', ' ').replace('</p>', ' ')
    regular = re.compile(r'[a-zA-Z]+://[^\s]*[.com|.cn][/*\S*]*')  
    content = regular.sub('', content_raw1)
    return content
```

```python
trans_list = [] #store the post that need to be translated
```

```python
for i in range(0, len(combine)): #detect whether the language of the post is English
    if combine.loc[i][32] == 'en':
        ori_nar = remove_tag(str(combine.loc[i][9]))
        combine.loc[i,'rtg_translate'] = ori_nar

    elif combine.loc[i][32] != 'en' and pd.isnull(combine.loc[i][32]) == False:  
        if combine.loc[i][32] == '':
            combine.loc[i,'rtg_translate'] = str(combine.loc[i][9])
        else:
            trans_list.append(i) #put the pist into trans_list if its languge is not English
```

```python
print(len(trans_list))
print(trans_list)
```

```python
for j in range(0, len(trans_list)):
    i = trans_list[j]
    txt = remove_tag(str(combine.loc[i][9])).lower()
    txt = find_regular(txt)
    #print(txt)
    if combine.loc[i][32] == '':
        combine.loc[i,'rtg_translate'] = str(combine.loc[i][9])
    elif len(txt) <= 120:                           #for short post, directly translate it
        combine.loc[i,'rtg_translate'] = translate.auto_from_buffer(txt,'en')
    else:                                           # for long post, chop it into small pieces
        txt = re.split('[\.\?!;\n\(\),৷\r\–]', txt)  #use punctuation to segment post 
        st = ''
        lan = ['sq','hi','bn','ml']                 #these language are easy to cause timeout errot
        for j in txt:
            if j != '':
                if len(j) > 120 and combine.loc[i][32] in lan:
                    k = 0
                    while k < len(j):               #chop the post
                        if k+100 >= len(j):
                            st += translate.auto_from_buffer(j[k:],'en')
                        else:
                            st += translate.auto_from_buffer(j[k: k+100],'en')
                        st += '.'
                        k += 100
                else:
                    st += translate.auto_from_buffer(j,'en')
                    st += '. '

        combine.loc[i,'rtg_translate'] = st
    print(i)
```

```python
combine.to_csv('combine_rtg.csv', index=False, encoding='utf-8')
```
