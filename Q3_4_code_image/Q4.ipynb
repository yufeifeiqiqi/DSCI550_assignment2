### Are there correlations between the sporting events, or the entertainment events with locations?

```python
print(df.columns)
```

```python
df['Sport']
```

```python
geo_entertainment_dic ={}
for i in range(len(df)):
    if pd.isnull(df['Film'][i]) ==False:
        entertainment = df['Film'][i]
    else:
        entertainment = 'NA'
          
    # check if geo_name column has geolocation names
    if pd.isnull(df['geo_name'][i])==False   :
        geo_infomration = df['geo_name'][i]
        geo_infomration = geo_infomration[2:-2]
        geo_infomration = geo_infomration.replace('\'',"")
        geo_name = geo_infomration.split(",") # a list of goe_names in each row
        for name in geo_name:
            name = name.strip()
            #print(name)
             
            if name not in geo_entertainment_dic:
                #print(i)
                geo_entertainment_dic[name] = {}
                geo_entertainment_dic[name][entertainment] = geo_entertainment_dic[name].get(entertainment, 0) + 1
            else:
                geo_entertainment_dic[name][entertainment] = geo_entertainment_dic[name].get(entertainment, 0) + 1
        #print(geo_infomration)
        #print(geo_name)
print(geo_entertainment_dic)
```

```python
df_geo_ent = pd.DataFrame.from_dict(geo_entertainment_dic)

df_geo_ent = df_geo_ent.replace(np.nan, 0)
df_geo_ent
```

```python
# fine the top 15 entertaments, including NA
geo_ent_index_top15 = df_geo_ent.sum(axis = 1).sort_values(ascending=False).head(15).index
geo_ent_index_top15 = list(geo_ent_index_top15)
for i in range(len(geo_ent_index_top15)):
    print(geo_ent_index_top15[i],'\n')
```

```python
df_geo_ent_2 = pd.DataFrame(df_geo_ent.T, columns = geo_ent_index_top15)

df_geo_ent_2
```

```python
# find the maximum value for each column
maxValueIndex = df_geo_ent_2.idxmax(axis = 'index')
#print("Maximum values of columns are at row index position :")
#print(maxValueIndex)
maxValueIndex
```

```python
for i in range(len(maxValueIndex)):
    print(geo_ent_index_top15[i], ' : ', maxValueIndex[i], df_geo_ent_2.loc[maxValueIndex[i],geo_ent_index_top15[i]],'\n')
    
```
