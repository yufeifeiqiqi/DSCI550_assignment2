```python
# lenght of parsed['metadata'] is 7 if no location mentioned
# one more location metioned, lenght +3 
geo_list = {}
for i in range(len(df['rtg_translate'])):

    #geo_name =[]
    if pd.isnull(df['rtg_translate'][i]):
        geo_list[i] = {}
        geo_list[i]['geo_name'] = "NA"
        geo_list[i]['long'] = "NA"
        geo_list[i]['lat'] = 'NA'
        continue
        
    #print(df['rtg_translate'][i])  
    parsed = parser.from_buffer(df['rtg_translate'][i], headers={'Content-Type' : 'application/geotopic'})
    #print(parsed)
    
    lenght = len(parsed['metadata'])
    #print(lenght)
    if 'Geographic_NAME' in parsed['metadata']:
        #print(parsed['metadata']['Geographic_NAME'])
        #print(parsed['metadata']['Geographic_LONGITUDE'])
        #print(parsed['metadata']['Geographic_LATITUDE'])
        geo_list[i] = {}
        geo_list[i]['geo_name'] = [] 
        geo_list[i]['geo_name'].append(parsed['metadata']['Geographic_NAME'])
        
        geo_list[i]['long'] = [] 
        geo_list[i]['long'].append(parsed['metadata']['Geographic_LONGITUDE'])
        
        geo_list[i]['lat'] = [] 
        geo_list[i]['lat'].append(parsed['metadata']['Geographic_LATITUDE'])
        
        # if count_location =1 only the geographic city. if greater than 1, wee need to find optional name for extra city
        count_location = int((lenght-7)//3)
        if count_location >0:
            for count in range(1, count_location):
                #print(count)
                geo_list[i]['geo_name'].append(parsed['metadata']['Optional_NAME' + str(count)])
                geo_list[i]['long'].append(parsed['metadata']['Optional_LONGITUDE'+str(count)])
                geo_list[i]['lat'].append(parsed['metadata']['Optional_LATITUDE'+ str(count)])
    else:
        geo_list[i] = {}
        geo_list[i]['geo_name'] = "NA"
        geo_list[i]['long'] = "NA"
        geo_list[i]['lat'] = 'NA'
        
        
        #print('NA')
#print(geo_list)
```
